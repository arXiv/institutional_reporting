```python
from bokeh.io import output_file, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.palettes import Blues8
import pandas as pd
from datetime import date
from bokeh.models import CheckboxGroup, CustomJS, DateRangeSlider
from bokeh.layouts import column


output_notebook()

# Read the CSV file into a pandas DataFrame
df2 = pd.read_csv('Arxiv Sample 2.csv')
# Get the unique values in the "Category Id" column
years = df2['Year'].unique().tolist()
years.sort()
#years = list(map(lambda x:str(x),years))

# Get the unique values in the "Category Id" column
papers = df2['Category Id'].unique().tolist()

for year in years:
    counts = df2[df2['Year'] == year]['Category Id'].value_counts()
    data[str(year)] = [counts.get(paper, 0) for paper in papers]

output_file("bars.html")

x = [ (paper, str(year)) for paper in papers for year in years ]
#counts = sum(zip(data['2019'], data['2020'], data['2021']), ()) #check to automate this process for further years
pap_len = len(papers)
counts = list()
for i in range (pap_len):
  for year in years:
    counts.append(data [str(year)][i])

source = ColumnDataSource(data=dict(x=x, counts=counts))
"""
Create a ColumnDataSource object for a stacked bar chart.

The ColumnDataSource object is created using a dictionary that maps column names to their corresponding data.
In this case, the dictionary has two keys: 'x' and 'counts'. The 'x' key maps to a list of tuples representing
combinations of paper categories and years. The 'counts' key maps to a flattened list containing the counts of papers
for each combination of paper category and year.

Parameters:
x : list of tuples
A list of tuples with each tuple representing a combination of a paper category and a year.
counts : list of integers
A flattened list containing the counts of papers for each combination of paper category and year.

Returns:
source : bokeh.models.sources.ColumnDataSource
A ColumnDataSource object that maps the column names ('x' and 'counts') to their respective data. This object
can be used in the vbar function to draw the bars in the plot.
"""

# Include institution name using HTML tags
p = figure(x_range=FactorRange(*x), height=250, title="Institution / Archive / Year",
           toolbar_location=None, tools="")
p.title.align = "center" 
p.title.text_font_size = "24px"

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="white",
       fill_color=factor_cmap('x', palette=Blues8, factors=[str(year) for year in years], start=1, end=2),legend = "Archive")

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1.6
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.major_tick_in = -6

# Stylize the plot area
p.plot_width = 700
p.plot_height = 300


# Legend

p.legend.location = "top_right"

LABELS = papers

checkbox_group = CheckboxGroup(labels=LABELS, active=[0, 1])
checkbox_group.js_on_click(CustomJS(code="""
    console.log('checkbox_group: active=' + this.active, this.toString())
"""))

show(checkbox_group)


date_range_slider = DateRangeSlider(value=(date(years[0], 1, 1), date(years[-1], 12, 31)),
                                    start=date(years[0], 1, 1), end=date(years[-1], 12, 31))
date_range_slider.js_on_change("value", CustomJS(code="""
    console.log('date_range_slider: value=' + this.value, this.toString())
"""))

show(date_range_slider)


output_notebook()
show(p)
```

```python
from bokeh.core.property_mixins import FontStyle
from bokeh.io import output_file, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange, LabelSet
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.palettes import Blues8
import pandas as pd
from datetime import date
from bokeh.models import CheckboxGroup, CustomJS, DateRangeSlider, RangeSlider
from bokeh.layouts import column
from bokeh.layouts import row
from bokeh.models import Div
import json

output_notebook()

# Read the CSV file into a pandas DataFrame
df2 = pd.read_csv("Arxiv Sample 2.csv")
# Get the unique values in the "Category Id" column
years = df2['Year'].unique().tolist()
years.sort()

# Get the unique values in the "Category Id" column
papers = df2['Category Id'].unique().tolist()

data = {}
for year in years:
    counts = df2[df2['Year'] == year]['Category Id'].value_counts()
    data[str(year)] = [counts.get(paper, 0) for paper in papers]

output_file("bars.html")

x = [(paper, str(year)) for paper in papers for year in years]
pap_len = len(papers)
counts = list()
for i in range(pap_len):
    for year in years:
        counts.append(data[str(year)][i])

source = ColumnDataSource(data=dict(x=x, counts=counts))

#p = figure(x_range=FactorRange(*x), height=250, title="Institution / Archive / Year",
#           toolbar_location=None, tools="")
p = figure(x_range=FactorRange(*x), height=250, title="Institution / Archive / Year",
           toolbar_location='below')


p.title.align = "center"
p.title.text_font_size = "24px"

p.vbar(x='x', top='counts', width=0.9, source=source, line_color="white",
       fill_color=factor_cmap('x', palette=Blues8, factors=[str(year) for year in years], start=1, end=2))

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1.6
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.major_tick_in = -6
p.yaxis.axis_label = "Papers"
p.yaxis.axis_label_text_font_style = 'bold'
p.yaxis.minor_tick_line_color = None

# Stylize the plot area
p.plot_width = 800
p.plot_height = 400

# Legend

LABELS = papers

checkbox_group = CheckboxGroup(labels=LABELS, active=list(range(len(papers))))

checkbox_group_all = CheckboxGroup(labels=["All"], active=[0])

callback_all = CustomJS(args=dict(checkbox_group_all=checkbox_group_all, checkbox_group=checkbox_group), code="""
    var count = checkbox_group.labels.length;
    console.log("All Clicked");
    if (checkbox_group_all.active.length == 0) {
        checkbox_group.active = Array();
    }
    else {
        checkbox_group.active = Array.from({length: count}, (_, index) => index);
    }
""")

range_slider = RangeSlider(
    title="Years",
    start=years[0],
    end=years[-1],
    step=1,
    value=(years[0], years[-1]))


# Convert the DataFrame to a JSON string and parse it as a JavaScript object
df2_json = json.dumps(df2.to_dict(orient='records'))
callback = CustomJS(args=dict(source=source, df2=df2_json, checkbox_group=checkbox_group, range_slider=range_slider, years=years, papers=papers), code="""
    var selected_categories = checkbox_group.active.map(i => checkbox_group.labels[i]);
    var start_date = new Date(range_slider.value[0], 1, 1);
    var end_date = new Date(range_slider.value[1], 1, 1);

    var df2 = JSON.parse(df2);

    var filtered_data = df2.filter(row => selected_categories.includes(row['Category Id']) &&
                                          row['Year'] >= start_date.getFullYear() &&
                                          row['Year'] <= end_date.getFullYear());
    
    var data = {};
    for (var year of years) {
        var counts = {};
        for (var row of filtered_data) {
                            if (row['Year'] == year) {
                    counts[row['Category Id']] = (counts[row['Category Id']] || 0) + 1;
                }
            }
        data[year] = counts;
    }

    var new_counts = [];
    for (var i = 0; i < papers.length; i++) {
        for (var year of years) {
            new_counts.push(data[year][papers[i]] || 0);
        }
    }

    source.data.counts = new_counts;
    source.change.emit();
""")

# Attach the callback to the CheckboxGroup and DateRangeSlider
checkbox_group.js_on_change("active", callback)
checkbox_group_all.js_on_click(callback_all)
range_slider.js_on_change("value", callback)

# Create an empty Div element with a specified width to create a gap
gap = Div(width=10)
legendTitle = Div(text="<b>Archives</b>")

# Create a layout with the plot, gap, and the CheckboxGroup and DateRangeSlider panels
layout = row(p, gap, column(legendTitle, checkbox_group_all, checkbox_group, range_slider))

output_notebook()
show(layout)
```
